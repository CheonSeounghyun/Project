```python
Traffic_fatal_accident_2015 = pd.read_csv('data/2015_Traffic_fatal_accident.csv',
                                                usecols = ['발생년월일시','주야','사망자수','사상자수','발생지시도',
                                                           '발생지시군구','법규위반','경도','위도'],
                                                encoding = 'euc-kr')
Traffic_fatal_accident_2016 = pd.read_csv('data/2016_Traffic_fatal_accident.csv',
                                                usecols = ['발생년월일시','주야','사망자수','사상자수','발생지시도',
                                                           '발생지시군구','법규위반','경도','위도'],
                                                encoding = 'euc-kr')
Traffic_fatal_accident_2017 = pd.read_csv('data/2017_Traffic_fatal_accident.csv',
                                                usecols = ['발생년월일시','주야','사망자수','사상자수','발생지시도',
                                                           '발생지시군구','법규위반','경도','위도'],
                                                encoding = 'euc-kr')
Traffic_fatal_accident_2015.head()
```

```python
Traffic_fatal_accident_sum = pd.concat([Traffic_fatal_accident_2012_2014,
                                    Traffic_fatal_accident_2015,
                                    Traffic_fatal_accident_2016,
                                    Traffic_fatal_accident_2017], ignore_index=True)
Traffic_fatal_accident_sum.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
```

```python
regions = list(set(Traffic_fatal_accident_sum["발생지시도"]))
Traffic_fatal_accident = pd.DataFrame(index=regions, 
                                      columns = ['총사고발생수','총사망자수','총사상자수',
                                                 '2012-2014','2015','2016','2017','평균사망자','평균사상자'])
```

```python
def summary(df,region,category,figure):
    # dataFrame, 지역, 컬럼, (평균,합계)
    if figure == "sum":
        result = df[df["발생지시도"]==region][category].sum()
    if figure == "mean":
        result = df[df["발생지시도"]==region][category].mean()
    return result
```

```python
for region in regions:
    Traffic_fatal_accident.loc[region]["총사망자수"] = summary(Traffic_fatal_accident_sum,region,"사망자수","sum")
    Traffic_fatal_accident.loc[region]["총사상자수"] = summary(Traffic_fatal_accident_sum,region,"사상자수","sum")
    Traffic_fatal_accident.loc[region]["평균사망자"] = summary(Traffic_fatal_accident_sum,region,"사망자수","mean")
    Traffic_fatal_accident.loc[region]["평균사상자"] = summary(Traffic_fatal_accident_sum,region,"사상자수","mean")
    Traffic_fatal_accident.loc[region]["2012-2014"] = len(Traffic_fatal_accident_2012_2014[Traffic_fatal_accident_2012_2014["발생지시도"]==region])
    Traffic_fatal_accident.loc[region]["2015"] = len(Traffic_fatal_accident_2015[Traffic_fatal_accident_2015["발생지시도"]==region])
    Traffic_fatal_accident.loc[region]["2016"] = len(Traffic_fatal_accident_2016[Traffic_fatal_accident_2016["발생지시도"]==region])
    Traffic_fatal_accident.loc[region]["2017"] = len(Traffic_fatal_accident_2017[Traffic_fatal_accident_2017["발생지시도"]==region])
    Traffic_fatal_accident.loc[region]["총사고발생수"] = len(Traffic_fatal_accident_sum[Traffic_fatal_accident_sum["발생지시도"]==region])
```
